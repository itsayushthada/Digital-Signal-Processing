```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def DFT(signal, N):
    k = np.linspace(0, N-1, N)
    img = np.zeros(N)
    real = np.zeros(N)
    for idx in range(N):
        img[idx]= np.sum(signal * np.sin(2*np.pi*idx*k/N))
        real[idx] = np.sum(signal * np.cos(2*np.pi*idx*k/N))
    
    return real,-img
```

**Question 6:** (Difficulty: ⋆⋆) Consider the three sinusoids of length $N=64$ as illustrated in the above figure; note that the signal values are shown from $n=0$ to $n=63$

Call $y_1[n]$ the blue signal, 
$y_2[n]$ the green
$y_3[n]$ the red.<br>
Further, define $x[n] = y_1[n] + y_2[n] + y_3[n]$ 
Note that the capital letters indicate the DFT vectors.

```python
N = 64

time = np.linspace(0, N-1, N)

y1 = 2*np.cos(2*np.pi*time/16)
y2 = 0.5*np.sin(2*np.pi*time/8)
y3 = np.ones_like(time)
ref_line = np.zeros(N)
x = y1 + y2 + y3

########################################

plt.rcParams["figure.figsize"] = (24, 6)
plt.plot(time, y1, color="blue", label="y1")
plt.plot(time, y2, color="green", label="y2")
plt.plot(time, y3, color="red", label="y3")
plt.plot(ref_line, color="black", label="Time")
plt.plot(time, x, "--", color='orange', label="New Signal")
plt.legend(loc = 4)
plt.show()
```

```python
np.linalg.norm(x, ord=2)**2
```

```python
real, img = DFT(x, N)

np.linalg.norm(np.sqrt(real**2 + img**2), ord=2)**2
```

```python
plt.plot(real.round(2))
plt.plot(img.round(2))
```
